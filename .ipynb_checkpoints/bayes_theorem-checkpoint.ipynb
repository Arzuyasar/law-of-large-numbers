```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Weather": ["Sunny","Sunny","Overcast","Rain","Rain","Rain","Overcast","Sunny","Sunny","Rain"],
    "Play":    ["No","No","Yes","Yes","Yes","No","Yes","No","Yes","Yes"]
}

df = pd.DataFrame(data)
df
```

```python
p_yes = len(df[df["Play"] == "Yes"]) / len(df)
p_no = len(df[df["Play"] == "No"]) / len(df)

p_yes, p_no
```

```python
p_sunny_yes = len(df[(df["Weather"] == "Sunny") & (df["Play"] == "Yes")]) / len(df[df["Play"] == "Yes"])
p_sunny_no  = len(df[(df["Weather"] == "Sunny") & (df["Play"] == "No")]) / len(df[df["Play"] == "No"])

p_sunny_yes, p_sunny_no

round(p_sunny_yes, 2), round(p_sunny_no, 2)
```

```python
p_sunny = len(df[df["Weather"] == "Sunny"]) / len(df)
p_rain = len(df[df["Weather"] == "Rain"]) / len(df)
p_overcast = len(df[df["Weather"] == "Overcast"]) / len(df)

p_sunny, p_rain, p_overcast
```

```python
p_play_yes_given_sunny = (p_sunny_yes * p_yes) / p_sunny
p_play_yes_given_sunny
round(p_play_yes_given_sunny, 2)
```

```python
p_rain_yes = len(df[(df["Weather"] == "Rain") & (df["Play"] == "Yes")]) / len(df[df["Play"] == "Yes"])

p_play_yes_given_rain = (p_rain_yes * p_yes) / p_rain
p_play_yes_given_rain
round(p_play_yes_given_rain, 2)
```

```python
print("Sunny iken maç oynanma olasılığı:", round(p_play_yes_given_sunny, 2))
print("Rain iken maç oynanma olasılığı:", round(p_play_yes_given_rain, 2))
```
