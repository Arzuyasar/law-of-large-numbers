```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 10000

dice = np.random.randint(1, 7, size=n)

cum_avg = np.cumsum(dice) / np.arange(1, n+1)

plt.plot(cum_avg)
plt.axhline(y=3.5, linestyle='--')
plt.title("Zar Deneyi - Ortalama Yakınsaması")
plt.xlabel("Deneme Sayısı")
plt.ylabel("Ortalama")
plt.show()
```

```python
coin = np.random.randint(0, 2, size=n)

cum_avg_coin = np.cumsum(coin) / np.arange(1, n+1)

plt.plot(cum_avg_coin)
plt.axhline(y=0.5, linestyle='--')
plt.title("Yazı-Tura Deneyi")
plt.xlabel("Deneme Sayısı")
plt.ylabel("Ortalama")
plt.show()
```

```python
binom = np.random.binomial(n=10, p=0.7, size=n)

cum_avg_binom = np.cumsum(binom) / np.arange(1, n+1)

plt.plot(cum_avg_binom)
plt.axhline(y=7, linestyle='--')
plt.title("Binom Deneyi")
plt.xlabel("Deneme Sayısı")
plt.ylabel("Ortalama")
plt.show()
```
